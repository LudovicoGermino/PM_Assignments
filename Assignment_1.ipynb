```python
# Import libraries
import yfinance as yf
import pandas as pd
import numpy as np

# Define the stock tickers and the date range
tickers = ["META", "AMZN", "AAPL", "NFLX", "GOOG"]
start_date = "2015-01-01"
end_date = "2020-12-31"

# Download daily stock prices
faang_data = yf.download(tickers, start=start_date, end=end_date)

# Save to a CSV file
faang_data.to_csv("faang_stock_data_2015_2020.csv")

# Display a preview of the data
faang_data.head()
```
